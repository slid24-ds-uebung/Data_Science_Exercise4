```bash
# b) .gitignore anlegen und to_be_ignored.txt aus dem Tracking entfernen
echo "to_be_ignored.txt" >> .gitignore   # Eintrag in die .gitignore schreiben
git add .gitignore

# --cached entfernt die Datei aus dem Index, behaelt sie aber lokal
git rm --cached to_be_ignored.txt
git commit -m "Add .gitignore and untrack to_be_ignored.txt"

git ls-files     # to_be_ignored.txt taucht nicht mehr auf
git status       # die Datei wird jetzt ignoriert
```

#### g) [2.5%]

 - Erstellen Sie einen Fork des remote Repositorys auf GitHub. **Dies muss durch einen anderen Benutzer geschehen, falls das original Repository nicht zu einer Organisation gehört!**

#### h) [10%]
- Der Pull-Request sollte anzeigen, dass es Konflikte gibt.

```bash
# h) Merge-Konflikt loesen: main in den Feature-Branch mergen
# Da main (f) und feature (c) dieselbe Datei mergesort.py geaendert
# haben, zeigt der Pull-Request einen Konflikt.
git switch feature/clean-mergesort
git fetch origin
git merge origin/main          # erzeugt Konfliktmarkierungen <<< === >>>

# mergesort.py oeffnen, Konfliktmarkierungen entfernen und die
# gewuenschte (bereinigte + Subplot-) Version behalten.
git add mergesort.py
git commit                     # Merge-Commit abschliessen

git push fork feature/clean-mergesort   # aktualisierten Branch pushen
# Auf GitHub: der PR ist nun konfliktfrei -> "Merge pull request".
```
